# Visualizing P-values in Different Ways

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Simulating p-value data for 54 tests across 3 different conditions
np.random.seed(42)
data = np.random.rand(54, 3)
df = pd.DataFrame(data, index=[f'Test {i+1}' for i in range(54)], columns=['Condition A', 'Condition B', 'Condition C'])
```

## 1. Clustered Heatmap

```python
# Clustered heatmap
plt.figure(figsize=(12, 10))
sns.clustermap(df, annot=True, fmt=".3f", cmap="YlGnBu", cbar_kws={'label': 'p-value'}, method='average')

# The clustermap itself handles the title within its axes
plt.title('Clustered P-values from 54 Different Tests')
plt.show()
```

## 2. Bar Plot of P-values for Condition A

```python
# Example of a bar plot for the first condition
plt.figure(figsize=(12, 6))
sns.barplot(x=df.index, y=df['Condition A'], palette="viridis")

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Add a threshold line at p=0.05
plt.axhline(0.05, color='red', linestyle='--')

# Add titles and labels
plt.title('P-values for Condition A')
plt.ylabel('P-value')

# Show the plot
plt.show()
```

## 3. Violin Plot of P-values

```python
# Violin plot showing the distribution of p-values for each condition
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, palette="Set2", inner="point")
plt.title('Distribution of P-values Across Conditions')
plt.ylabel('P-value')
plt.show()
```

## 4. Box Plot of P-values

```python
# Box plot showing the distribution of p-values for each condition
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, palette="Set3")
plt.title('Box Plot of P-values Across Conditions')
plt.ylabel('P-value')
plt.show()
```

## 5. Swarm Plot of P-values

```python
# Swarm plot showing individual p-values for each condition
plt.figure(figsize=(10, 6))
sns.swarmplot(data=df, palette="Set1")
plt.title('Swarm Plot of P-values Across Conditions')
plt.ylabel('P-value')
plt.show()
```
